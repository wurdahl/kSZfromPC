```python
import healpy as hp
import numpy as np
import matplotlib.pyplot as plt
```

```python
maplensedOver = np.load('./MAPS/lensedOverdensity/_NS_1536_R_2968_P_1536_DV_192_Sd_336244_SB_5.npy')
mapOver = np.load('./MAPS/overdensity/_NS_1536_R_2968_P_2048_DV_128_Sd_995225_SB_7.npy')
```

```python
hp.mollview(maplensedOver)
```

```python
maplensedOver = maplensedOver/np.mean(maplensedOver)-1
```

```python
hp.mollview(mapOver)
```

```python
psOver = hp.anafast(mapOver)
pslensedOver = hp.anafast(maplensedOver)
```

```python
cross = hp.anafast(mapOver,maplensedOver)
```

```python
fig = plt.figure(dpi=300)
plt.loglog(psOver)
plt.loglog(pslensedOver)
plt.legend(['overdensity','lensedOverdensity'])
```

```python
plt.figure(dpi=300)
plt.plot(cross/(np.sqrt(np.multiply(psOver,pslensedOver))))
```

```python
fig = plt.figure(dpi=300)
plt.loglog(psOver/pslensedOver)
```
