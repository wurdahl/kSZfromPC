```python
import healpy as hp
import matplotlib.pyplot as plt
import numpy as np
```

```python
lensedCorr = np.load('./Lens_NS_1024_R_2968_P_2048_DV_192_SuperBin_5.npy')
notlensedCorr = np.load('./noLens_NS_1024_R_2968_P_2048_DV_192_SuperBin_5.npy')
```

```python
fig = plt.figure(dpi=400)
plt.semilogx(lensedCorr)
plt.semilogx(notlensedCorr)
plt.title('Correlation Between true velocity and velocity reconstruction');
plt.legend(['With lensing','Without lensing']);
plt.xlabel('L');
plt.ylabel('Correlation');
```

```python
lensedOverdensity = np.load('./lensedOverdensity_NS_1024_R_2968_P_2048_DV_192_SuperBin_5.npy')
overdensity = np.load('./overdensity_NS_1024_R_2968_P_2048_DV_192_SuperBin_5.npy')
```

```python
fig = plt.figure(dpi=400)
plt.loglog(lensedOverdensity)
plt.loglog(overdensity)
plt.title('Power Spectrum of Overdensity');
plt.legend(['With lensing','Without lensing']);
plt.ylim([10**-8,10**-4])
plt.xlabel('L');
plt.ylabel('Power');
```
