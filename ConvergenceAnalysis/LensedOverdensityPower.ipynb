```python
import healpy as hp
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
lensedMap = np.load('./MAPS/'+os.listdir('./MAPS')[1])
Map = np.load('./MAPS/'+os.listdir('./MAPS')[2])
```

```python
plt.hist(lensedMap, bins=np.linspace(0,30));
```

```python
hp.mollview(lensedMap);
```

```python
hp.mollview(Map,min=-1,max=5);
```

```python
np.mean(Map)
```

```python
np.mean(lensedMap)
```

```python
overdensityPS = hp.anafast(Map)
```

```python
lensedPS = hp.anafast(lensedMap)
```

```python
lensedMapConverted = (lensedMap-np.mean(lensedMap))/np.mean(lensedMap)
```

```python
lensedMapConvertedPS = hp.anafast(lensedMapConverted)
```

```python
plt.figure(dpi=400)
plt.loglog(lensedPS)
plt.loglog(overdensityPS)
plt.loglog(lensedMapConvertedPS)
plt.legend(['Original Lensed','Regular Overdensity','Converted Lensed Map'])
```

```python
plt.figure(dpi=400)
plt.loglog(overdensityPS)
plt.loglog(lensedMapConvertedPS)
plt.legend(['Regular Overdensity','Converted Lensed Map'])
```
